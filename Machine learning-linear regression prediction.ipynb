```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("canada_per_capita_income.csv")
```

```python
df
```

```python
df.head()
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('income')
plt.scatter(df.year,df.income,color='blue',marker='*')
```

```python
new_df=df.drop('income', axis='columns')
new_df
```

```python
income=df.income
income
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,income)
```

####  Predict canada's per capita income in year 2020

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
